# Matrix Operations

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
```

```python
inv_A = np.linalg.inv(A)
print(inv_A)
```

```python
proof1 = np.allclose(np.dot(A, inv_A), np.eye(2))
print(proof1)
```

```python
# Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
proof2 = not np.array_equal(AB, BA)

print(proof2)
```

```python
proof3 = np.allclose((np.dot(A, B)).T, np.dot(B.T, A.T))
print(proof3)
```

```python
coeff_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
print(coeff_matrix)
```

```python
const_vector = np.array([-1, -3, 9])
print(const_vector)
```

```python
solution = np.linalg.solve(coeff_matrix, const_vector)
print(solution)
```

# How fast numpy is?

```python
size = 1000000
```

```python
list_a = list(range(size))
list_b = list(range(size))
array_a = np.arange(size)
array_b = np.arange(size)
```

```python
start_time = time.time()
result_list = [list_a[i] + list_b[i] for i in range(size)]
python_add_time = time.time() - start_time
print(f"Python List Addition: {python_add_time:.5f} seconds")
```

```python
start_time = time.time()
result_array = array_a + array_b
numpy_add_time = time.time() - start_time
print(f"NumPy Array Addition: {numpy_add_time:.5f} seconds")
```

```python
# Python List Multiplication
start_time = time.time()
result_list = [list_a[i] * list_b[i] for i in range(size)]
python_mul_time = time.time() - start_time
print(f"Python List Multiplication: {python_mul_time:.5f} seconds")
```

```python
# NumPy Array Multiplication
start_time = time.time()
result_array = array_a * array_b
numpy_mul_time = time.time() - start_time
print(f"NumPy Array Multiplication: {numpy_mul_time:.5f} seconds")
```

```python
# Python List Dot Product
start_time = time.time()
dot_product_list = sum(list_a[i] * list_b[i] for i in range(size))
python_dot_time = time.time() - start_time
print(f"Python List Dot Product: {python_dot_time:.5f} seconds")
```

```python
# NumPy Array Dot Product
start_time = time.time()
dot_product_array = np.dot(array_a, array_b)
numpy_dot_time = time.time() - start_time
print(f"NumPy Array Dot Product: {numpy_dot_time:.5f} seconds")
```

```python
matrix_size = 1000
mat_a = np.random.rand(matrix_size, matrix_size)
mat_b = np.random.rand(matrix_size, matrix_size)
list_mat_a = mat_a.tolist()
list_mat_b = mat_b.tolist()

# NumPy Matrix Multiplication
start_time = time.time()
result_numpy_mat = np.matmul(mat_a, mat_b)
numpy_mat_time = time.time() - start_time
print(f"NumPy Matrix Multiplication: {numpy_mat_time:.5f} seconds")
```
